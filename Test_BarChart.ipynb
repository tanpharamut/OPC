```python
labels = ['ความพึงพอใจ\nของลูกค้า', 'ความไม่พึงพอใจ\nของลูกค้า', 'ความผูกพันที่มีต่อ\nผลิตภัณฑ์','ภาพรวม']
y = [78.6, 50.4, 81.3, 70.11]

x = np.arange(len(labels))  # the label locations
width = 0.35

# Create a figure and axis object
fig, ax = plt.subplots()

# Plot the data as bars on the axis object
ax.bar(x, y, width)


# Set the chart title and labels for the x and y axes
ax.set_title('ทดสอบ')
ax.set_xlabel('X-axis label')
ax.set_ylabel('Y-axis label')
ax.set_xticks(x)
ax.set_xticklabels(labels)

# Show the chart
plt.hist(0.75,facecolor = 'violet', alpha = 0.75)
plt.show()
```

```python
# Create some sample data
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

# Create the plot
fig, ax = plt.subplots()
ax.plot(x, y)

# Adjust the figure size
fig.set_size_inches(10, 6)

# Rotate the x-axis labels
plt.xticks(rotation=45)

# Reduce the font size
plt.rcParams.update({'font.size': 12})

# Increase the spacing between x-axis labels
plt.xticks(x, ['ความพึงพอใจของลูกค้า', 'ความไม่พึงพอใจของลูกค้า', 'ความผูกพันที่มีต่อผลิตภัณฑ์','ภาพรวม'])

# Add axis labels and a title
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_title('Graph Title')

plt.show()
```
